```python
import joblib
import pandas as pd

# Load your model
model = joblib.load('D:\clv\models\linreg.pkl')

def predict_clv(total_qty, avg_unit_price, monetary):
    # Map columns in the EXACT order your CSV used for training
    customer_data = pd.DataFrame({
        'TotalQty': [total_qty],
        'AvgUnitPrice': [avg_unit_price],
        'Monetary_Value': [monetary]
    })
    
    raw_prediction = model.predict(customer_data)[0]
    
    # Apply the floor: Nobody can have a negative CLV
    final_clv = max(0, raw_prediction)
    
    return final_clv

# Test the luxury buyer again
luxury_clv = predict_clv(total_qty=5, avg_unit_price=450.00, monetary=2250.00)
print(f"Luxury Buyer 3-Month CLV: ${luxury_clv:.2f}")
```

```python
# Print the starting baseline (Intercept)
print(f"Base CLV (Intercept): ${model.intercept_:.2f}")

# Print the weights (Coefficients)
print(f"Weight for Total Qty: {model.coef_[0]:.2f}")
print(f"Weight for AvgUnitPrice: {model.coef_[1]:.2f}")
print(f"Weight for Monetary:    {model.coef_[2]:.2f}")
```
